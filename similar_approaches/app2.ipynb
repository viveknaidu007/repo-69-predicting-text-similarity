```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer  # Import for stemming
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re

# Load data (replace with your actual loading method)
data = pd.read_csv("DataNeuron_Text_Similarity.csv")
paragraph1 = data["text1"]
paragraph2 = data["text2"]

# Preprocess text data
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove punctuation
    stop_words = stopwords.words('english')  # Load stop words from NLTK
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stop words
    stemmer = PorterStemmer()  # Create a PorterStemmer object
    text = ' '.join([stemmer.stem(word) for word in text.split()])  # Apply stemming
    return text

paragraph1_clean = paragraph1.apply(preprocess_text)
paragraph2_clean = paragraph2.apply(preprocess_text)

# Feature extraction with TF-IDF
vectorizer = TfidfVectorizer()
features_para1 = vectorizer.fit_transform(paragraph1_clean)
features_para2 = vectorizer.transform(paragraph2_clean)

# Calculate semantic similarity using cosine similarity
similarity_scores = cosine_similarity(features_para1, features_para2)

# Print the first 5 pairs of paragraphs and their similarity scores
for i in range(5):
    print("Paragraph 1:", paragraph1.iloc[i])
    print("Paragraph 2:", paragraph2.iloc[i])
    print("Similarity Score:", similarity_scores[i][0])
    print("-" * 50)
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from gensim.models import Word2Vec
import numpy as np
import re

# Load data (replace with your actual loading method)
data = pd.read_csv("DataNeuron_Text_Similarity.csv")
paragraph1 = data["text1"]
paragraph2 = data["text2"]

# Preprocess text
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove punctuation
    stop_words = stopwords.words('english')
    tokens = word_tokenize(text)
    tagged_tokens = pos_tag(tokens)  # Perform POS tagging
    filtered_tokens = [word for word, pos in tagged_tokens if pos not in ['DT', 'PRP', 'IN']]  # Remove specific POS categories
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text

paragraph1_clean = paragraph1.apply(preprocess_text)
paragraph2_clean = paragraph2.apply(preprocess_text)

# Download and load pre-trained Word2Vec model
model = Word2Vec.load('word2vec-google-news-300')  # Download and load Google News model (300-dimensional vectors)

# Function to get word embeddings from a sentence
def get_embeddings(text):
    embeddings = []
    for word in text.split():
        try:
            embeddings.append(model.wv[word])
        except KeyError:  # Word not found in vocabulary
            embeddings.append(np.zeros(model.vector_size))  # Use zero vector for missing words
    return np.mean(embeddings, axis=0)  # Average word vectors

# Get word embeddings for paragraphs
paragraph1_embedding = get_embeddings(paragraph1_clean.iloc[0])
paragraph2_embedding = get_embeddings(paragraph2_clean.iloc[0])

# Calculate weighted cosine similarity
def weighted_cosine_similarity(embedding1, embedding2, weights):
    similarity = np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))
    weighted_similarity = similarity * np.sum(weights)
    return weighted_similarity

# Define weights based on POS categories (adjust as needed)
weights = {"NN": 2, "JJ": 1.5, "VB": 1}  # Higher weight for nouns, adjectives, and verbs

similarity_score = weighted_cosine_similarity(paragraph1_embedding, paragraph2_embedding, weights)

# Print similarity score
print("Similarity Score:", similarity_score)
```
